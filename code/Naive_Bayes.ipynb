<h1>Ejemplos de de Nayve Bayes</h1>

```python
from sklearn.datasets import make_classification #Se importa desde una paquetería make_classification para generar datos con clasificación

X, y = make_classification( #Se generan los datos clasificados de entrenamiento con los siguientes parámetros:
    n_features=6, #Cada clase tiene 6 carácteristicas
    n_classes=3, #Hay 3 clases
    n_samples=800, #Se usan 800 muestras

    random_state=1, #fijar una semilla
    n_clusters_per_class=1 
)
y[:3] # se imprimen unos elementos etiquetados para ver que las clases son los números 0,1 y 2
```

```python
X
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, marker="*")
```

```python
from sklearn.model_selection import train_test_split
#Se parten los datos de manera aleatoria de acuerdo al porcentaje indicado, para obtener datos para test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
from sklearn.naive_bayes import GaussianNB #Se importa naive bayes gausiano

#crea el clasificador
model = GaussianNB()

##entrena el modelo
model.fit(X_train, y_train)

#predice
predicted = model.predict(X_test)

predicted
```

```python
#Podemos ver las probabilidades que asignó el modelo para elegir la clase a la que pertenece cada dato
predicted_prob=model.predict_proba(X_test)
#Por ejemplo para el tercer dato, se eligió la categoria 2, pues tiene probabilidad 1
```

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
)
#Para evaluar en los datos test, obtenemos las métricas siguientes
y_pred = model.predict(X_test)
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

<h3> Naive bayes para categorias</h3>

```python
import numpy as np
rng = np.random.RandomState(1)
X = rng.randint(5, size=(1000, 100)) #Se generan 6 datos con 100 caracteristicas
y = rng.randint(1,7,size=(1000)) #sus variables de respuesta son del 1 al 6
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125,stratify=y
)
```

```python
from sklearn.naive_bayes import CategoricalNB #Importa Naive bayes
clf = CategoricalNB()
clf.fit(X_train, y_train)
predict=clf.predict(X_test)
```

```python

accuray = accuracy_score(predict, y_test)
f1 = f1_score(predict, y_test, average="weighted")
accuray
```
