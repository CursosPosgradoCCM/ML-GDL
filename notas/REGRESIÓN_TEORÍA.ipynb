## 1. Definición de Regresión
El **análisis de regresión** engloba a un conjunto de métodos estadísticos que usamos cuando tanto la **variable
de respuesta** como  la(s) **variable(s) predictiva(s)** son contínuas y queremos predecir valores de la
primera en función de valores observados de las segundas.

En esencia, el análisis de regresión consiste en **ajustar un modelo a los datos**, **estimando coeficientes** a partir de las observaciones, con el fin de **predecir valores** de la **variable de respuesta** a partir de una (regresión simple) o más **variables** (regresión múltiple)
**predictivas** o explicativas.

## 2. Usos principales de la regresión:
   * **Predicción:** La regresión **ajusta un modelo a los datos** históricos y usa ese modelo para **hacer predicciones sobre nuevos datos**.

        *Ejemplo:* En el análisis de viviendas, puedes usar la regresión para predecir el precio de una casa basándote en características
                   como el número de habitaciones, el tamaño del lote, y la ubicación.
   * **Comprender Relaciones:** La regresión te permite ver cómo **cambian los resultados** cuando **cambian las variables predictoras**.

       _Ejemplo:_ Si eres un investigador de salud, puedes usar la regresión para entender cómo factores como la dieta y el ejercicio afectan la presión arterial.
   * **Identificar tendencias:**  La regresión ayuda a **modelar y visualizar tendencias** en los **datos.**

      _Ejemplo:_ Las empresas pueden usar la regresión para analizar las tendencias en las ventas a lo largo del tiempo y entender cómo factores como la estacionalidad afectan las ventas.
   * **Optimización:** Al modelar cómo las diferentes estrategias afectan los resultados, puedes tomar decisiones informadas sobre cómo  asignar recursos.

      _Ejemplo:_ Las empresas pueden utilizar la regresión para optimizar el gasto en marketing, al identificar qué canales publicitarios tienen el mayor impacto en las ventas.

## 3. Tipos de regresión:
A continuación se enlistan algunos de los tipos de regresión más utilizados:
| Tipos de regresión | Uso típico |
| --------- | --------- | 
| lineal simple | Predicción de **una variable de respuesta** cuantitativa a partir de **una variable predictora** cuantitativa |
| polinomial | Predicción de una **variable de respuesta** cuantitativa a partir de una **variable predictora** cuantitativa, donde la relación se modela como una **función polinomial** de orden n |
| lineal múltiple | Predicción de una **variable de respuesta** cuantitativa a partir de **dos o más variables predictoras** cuantitativas |
| multivariada | Predicción de **más de una variable de respuesta** cuantitativa a partir de **una o más variables predictoras** cuantitativas |
| logística | Predicción de **una variable categórica** a partir de **una o más predictoras** |
| de Poisson | Predicción de **una variable de respuesta que representa un conteo** a partir de **una o más predictoras** |
| no lineal | Predicción de **una variable de respuesta** cuantitativa a partir de **una o más predictoras**, donde **el modelo no es lineal**  |
| robusta |Predicción de **una variable de respuesta** cuantitativa a partir de **una o más predictoras**, usando una **aproximación resistente al efecto de observaciones influyentes**  |

Otro tipo de regresiones menos comunes: 

**Regresión Ridge o Lasso** (gran número de variables independientes, y es posible que algunas de ellas sean irrelevantes o que haya multicolinealidad (correlación entre variables independientes)).
    Ideal para: Mejorar la estabilidad del modelo y manejar la multicolinealidad. Ridge es útil cuando se desea regularizar el modelo sin eliminar variables, mientras que Lasso puede realizar selección de variables al llevar algunos coeficientes a cero.

**Regresión de Datos de Supervivencia o Análisis de Cox:** cuando estás interesado en el tiempo hasta que ocurre un evento (por ejemplo, tiempo hasta la falla de un componente).
    Ideal para: Modelar y analizar datos de tiempo de vida y eventos, como el tiempo hasta la recuperación de una enfermedad.

**Regresión de Series Temporales:** Cuando: Los datos tienen una estructura temporal y el objetivo es modelar la dinámica a lo largo del tiempo.
    Ideal para: Predecir valores futuros basados en valores pasados, como pronosticar las ventas mensuales de una empresa.

Elegir el modelo de regresión adecuado implica considerar la **naturaleza de tus datos**, el **objetivo del análisis** y los supuestos subyacentes de cada técnica.


## 3. Modelado estadístico: selección de modelos, ajuste a los datos

Dadas **dos variables cuantitativas** existen **muchos de modelos** que podrían **describir la relación matemática entre ellas**. 
El reto está en elegir el modelo que **mejor se ajuste a estos datos** para **minimizar el error** en la estima que se haga a partir del modelo.

Usamos los modelos para **estimar el valor promedio de la variable de respuesta** en función de parámetros estimados de los datos. De manera general, podemos predecir valores de  la variable de respuesta usando esta **fórmula**:
$$estima_i = (modelo) + error_i$$
_**En Machine Learning, el concepto de error se refiere a la discrepancia entre las predicciones del modelo y los valores reales (o verdaderos). Entender y manejar el error es crucial para construir modelos predictivos efectivos._

1. elegir una familia de modelos adecuados a nuestros datos (modelos lineales, polinomiales, exponenciales,no lineales . . . ),
2. determinar el grado de parametrización adecuado del modelo
3. obtener estimas de máxima verosimilitud de dichos parámetros

## 5. Mínimos cuadrados

El método de mínimos cuadrados es una **técnica utilizada para encontrar la mejor aproximación de un modelo a un conjunto de datos observados**. El objetivo es **minimizar la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo**.

En otras palabras, los mínimos cuadrados buscan ajustar el modelo de manera que la diferencia entre los valores reales y los valores predichos sea lo más pequeña posible en términos de suma de cuadrados.

### Relación entre minimos cuadrados y regresión lineal (simple)
En el contexto de la regresión lineal, el método de mínimos cuadrados se utiliza para ajustar una **línea recta** a los datos. 

Para la regresión lineal simple, el modelo tiene la forma:
   $$ y = \beta_0+\beta_1x+\epsilon $$

Donde:
* $y$ es la variable dependiente.
* $x$ es la variable independiente.
* $\beta_0$ es la intersección (o término constante) de la línea.
* $\beta_1$ es la pendiente de la línea.
* $\epsilon$ es el término de error o residuo.

Para encontrar los valores óptimos de $\beta_0$ y $\beta_1$ que mejor se ajusten a los datos, se usa el método de mínimos cuadrados. Esto implica minimizar la suma de los cuadrados de los errores residuales:
$$\text{SSE} = \sum_{i=1}^{n} \epsilon_i^2 = \sum_{i=1}^{n} \left[ y_i - (\beta_0 + \beta_1 x_i) \right]^2$$

Los valores $\beta_0$ y $\beta_1$ que minimizan la SSE se encuentran resolviendo un sistema de ecuaciones derivado de la minimización. Estos valores proporcionan la mejor línea de ajuste según el criterio de mínimos cuadrados.



De manera breve...

Para encontrar el mínimo de SSE, tomamos las derivadas parciales de SSE con respecto a $\beta_0$ y $\beta_1$, y las igualamos a cero.
$$ \frac{\partial \text{SSE}}{\partial \beta_0} = -2 \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right),$$
$$ \frac{\partial \text{SSE}}{\partial \beta_1} = -2 \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right) x_i$$
Igualando a cero,
$$ -2 \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right) = 0 $$
$$-2 \sum_{i=1}^{n} \left( y_i - \beta_0 - \beta_1 x_i \right) x_i = 0 $$
Resolviendo primero para $\beta_0$ y despues resolviendo para $\beta_1$ (sustituyendo $\beta_0$):
$$\beta_0 = \bar{y} - \beta_1 \bar{x}, $$
$$ \beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2} = \frac{S_{xy}}{S_{xx}}$$
Finalmente, sustituyendo $\beta_1$ en la ecuación de $\beta_0$:
$$ \beta_0 = \bar{y} - \beta_1 \bar{x} .$$

#### Para el caso de regresión polinomial y ajuste de rectas...
En el ajuste de datos usando un polinomio de grado $k$, el modelo tiene la forma:


$y_i = \beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \cdots + \beta_k x_i^k + \epsilon_i$

donde:

* $y_i$ es el valor observado de la variable dependiente.
* $x_i$ es el valor de la variable independiente.
* $\beta_k$ son los coeficientes del polinomio.
* $\epsilon_i$ es el término de error.
  
La suma de los cuadrados de los errores (SSE) para un modelo polinómico se define como:
$$ \text{SSE} = \sum_{i=1}^{n} \left[ y_i - \left( \beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \cdots + \beta_k x_i^k \right) \right]^2.$$
Para este caso se repite la historia, para encontrar los valores óptimos de $\beta_j$ que minimizan la SSE, tomamos las derivadas parciales de SSE con respecto a cada uno de los coeficientes $\beta_j$ y las igualamos a cero. 

#### Regresión múltiple y minimos cuadrados
En la regresión lineal múltiple, donde hay múltiples variables independientes, el modelo tiene la forma:


$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p + \epsilon$$

donde:

* $y$ es el valor observado de la variable dependiente.
* $x_i$ son las variables independientes.
* $\beta_0$ es el término constante (intersección).
* $\beta_j$ con $j\neq 0$ son los coeficientes de las variables independientes.
* $\epsilon$ es el término de error.

El objetivo del método de mínimos cuadrados es encontrar el vector de coeficientes $\beta= (\beta_0,..\beta_p)$ que minimiza la suma de los cuadrados de los errores. La suma de los cuadrados de los errores (SSE)  en este caso se define como:

$$ \text{SSE} = \sum_{i=1}^{n} \left[ y_i - \left( \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \cdots + \beta_p x_{ip} \right) \right]^2$$

Donde $x_{ij}$ son los valores de las variables independientes para la observación $i$ .Para encontrar los valores óptimos de los coeficientes, tomamos las derivadas parciales de SSE con respecto a cada $\beta_1,\beta_j$ y las igualamos a cero. Esto da lugar a un sistema de ecuaciones lineales que se puede resolver para obtener los valores óptimos de los coeficientes.

En términos matriciales, el problema se puede expresar como:
$$\mathbf{X} \boldsymbol{\beta} = \mathbf{y}+\epsilon$$

donde:
* $X$ es la matriz de diseño que contiene las variables independientes (incluyendo una columna de 1's para el término constante $\beta_0$).
* $\beta$ es el vector de coeficientes.
* $Y$ es el vector de valores observados de la variable dependiente.

La solución de mínimos cuadrados se encuentra resolviendo:

$$\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}.$$
*** _$X^{+}= (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T$ es conocida como la pesudoinversa, aún más la solución de nuestro problema esta dada por $\boldsymbol{\beta} = X^{+} \mathbf{y}.$_
*** $||X⁺||_2 ||X||_2= cond(X)$.

*** _Para resolver este problema existen muchos métodos númericos entre ellos: factorizacion QR(Householder,Givens, Gram-Smidth,SVD))_.

### 6. EJEMPLO:
Supongamos que tenemos un conjunto de datos con las siguientes columnas:
* **Horas Estudio**: El número de horas que un estudiante ha estudiado.
* **Nota**: La nota obtenida en un examen.
  | Horas Estudio | Nota |
|---------------|------|
| 1             | 50   |
| 2             | 55   |
| 3             | 60   |
| 4             | 65   |
| 5             | 70   |

Queremos ajustar un modelo de regresión lineal simple de la forma:
$$Nota = \beta_0 + \beta_1 \cdot Horas \text{ Estudio} + \epsilon$$
donde:

$\beta_0$ es el término constante.

$\beta_1$ es el coeficiente de Horas Estudio.
La matriz de diseño $X$ incluirá una columna para el término constante (1s) y una columna para Horas Estudio:
$$ \mathbf{X} =
\begin{bmatrix}
1 & 1 \\
1 & 2 \\
1 & 3 \\
1 & 4 \\
1 & 5
\end{bmatrix}
$$
 Y el vector de valores observados $y$ será:
 $$ \mathbf{y} =
\begin{bmatrix}
50 \\
55 \\
60 \\
65 \\
70
\end{bmatrix}
$$
Usando la fórmula de mínimos cuadrados:

$$ \boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

Primero, calculamos $X^TX$:

$$ \mathbf{X}^T \mathbf{X} =
\begin{bmatrix}
5 & 15 \\
15 & 55
\end{bmatrix} $$
Luego, calculamos $X^Ty$:

$$ \mathbf{X}^T \mathbf{y} =
\begin{bmatrix}
300 \\
1100
\end{bmatrix}$$

Ahora, resolvemos para $\beta$:

$$\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

Esto implica encontrar la inversa de $X^TX$ y multiplicarla por $X^Ty$ .
Después de realizar estos cálculos, obtenemos los coeficientes:
$$
\boldsymbol{\beta} =
\begin{bmatrix}
45 \\
5
\end{bmatrix}
$$
Esto significa que el modelo ajustado es:

$$ Nota = 45 + 5 \cdot Horas \text{ Estudio}$$

Interpretación:

$\beta_0 = 45$ La nota base cuando Horas Estudio es cero.

$\beta_1 = 5$: El aumento en la nota por cada hora adicional de estudio.

### Código en Python

El siguiente código realiza la regresión, hace la predicción y grafica los datos junto con la recta de ajuste.

Resultados:

El código imprimirá los coeficientes del modelo y la predicción para 6 horas de estudio, y mostrará una gráfica con los datos y la recta de ajuste. Aquí está el resultado esperado:

* Coeficientes: $\beta_0 = 45$ y  $\beta_1 = 5$
* Predicción: Para 6 horas de estudio, la nota predicha será 75.

La gráfica te permitirá visualizar cómo la recta de ajuste se ajusta a los datos observados.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame
data = {
    'Horas Estudio': [1, 2, 3, 4, 5],
    'Nota': [50, 55, 60, 65, 70]
}
df = pd.DataFrame(data)

# Definir la matriz de diseño X y el vector de valores observados y
X = df[['Horas Estudio']]
X = np.c_[np.ones(X.shape[0]), X]  # Añadir columna de 1s para el término constante,  La función np.c_ 
#permite combinar arrays de manera horizontal, )).es decir, apilar arrays en columnas
y = df['Nota']

# Calcular los coeficientes usando la fórmula de mínimos cuadrados
X_transpose = X.T
beta = np.linalg.inv(X_transpose @ X) @ X_transpose @ y

# Extraer los coeficientes
beta_0, beta_1 = beta
print(f"Coeficientes: beta_0 = {beta_0}, beta_1 = {beta_1}")

# Hacer una predicción
horas_estudio_nuevo = 6
nota_predicha = beta_0 + beta_1 * horas_estudio_nuevo
print(f"Predicción para {horas_estudio_nuevo} horas de estudio: {nota_predicha}")

# Graficar los datos y la recta de ajuste
plt.figure(figsize=(10, 6))

# Datos
plt.scatter(df['Horas Estudio'], df['Nota'], color='blue', label='Datos Observados')

# Recta de ajuste
x_values = np.linspace(df['Horas Estudio'].min(), df['Horas Estudio'].max(), 100)
y_values = beta_0 + beta_1 * x_values
plt.plot(x_values, y_values, color='red', label='Recta de Ajuste')

# Personalizar la gráfica
plt.xlabel('Horas de Estudio')
plt.ylabel('Nota')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.grid(True)
plt.show()
```
