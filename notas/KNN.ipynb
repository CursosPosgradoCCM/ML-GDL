 ## 1.¿Qué es KNN? 

El algoritmo de k vecinos más cercanos, también conocido como KNN o k-NN, es un clasificador de **aprendizaje supervisado** **no paramétrico**, que utiliza la proximidad para hacer **clasificaciones** o predicciones sobre la agrupación de un punto de datos individual.

Si bien se puede usar para problemas de **regresión** o clasificación, generalmente se usa como un algoritmo de clasificación, partiendo de la suposición de que se pueden encontrar puntos similares cerca uno del otro

```python
En machine learning, un **algoritmo no paramétrico** es un tipo de modelo que no hace suposiciones 
explícitas sobre la forma funcional de la distribución de los datos. En otras palabras, 
**no se define un conjunto fijo de parámetros** a priori que limite la estructura del modelo.
En lugar de ello, los algoritmos no paramétricos se adaptan directamente a los datos, y su complejidad 
puede aumentar a medida que se dispone de más datos.

### Características principales:
1. **No hay un número fijo de parámetros**:
   - A diferencia de los modelos paramétricos (como la regresión lineal, donde solo ajustas un 
    número fijo de coeficientes), los modelos no paramétricos no tienen un conjunto fijo de parámetros.
   - A medida que aumenta la cantidad de datos, el modelo puede ajustarse de manera más compleja y 
    aumentar el número de "parámetros" implícitos (aunque no se llamen así formalmente).

2. **Más flexibilidad**:
   - Los modelos no paramétricos pueden adaptarse mejor a la estructura compleja de los datos sin
estar restringidos por una forma predeterminada. Esto los hace útiles cuando no se conoce la estructura 
subyacente de los datos o cuando se espera que sea muy compleja.

3. **Mayor riesgo de sobreajuste (overfitting)**:
   - Dado que estos modelos tienen la capacidad de adaptarse muy bien a los datos,
también corren el riesgo de sobreajustarse a los detalles y ruido de los datos de entrenamiento, 
lo que puede reducir su capacidad de generalización en datos nuevos.

### Ejemplos de algoritmos no paramétricos:

1. **k-Nearest Neighbors (k-NN)**:
   - Este algoritmo clasifica un nuevo punto basado en los puntos más cercanos en el conjunto de entrenamiento. No asume una forma funcional para la frontera de decisión y, en cambio, depende completamente de los datos.
   - No tiene un modelo explícito: simplemente almacena los datos y realiza las predicciones en
       función de ellos.

2. **Árboles de decisión**:
   - Los árboles de decisión dividen repetidamente los datos en función de características,
       lo que les permite capturar relaciones no lineales. La estructura del árbol no está predefinida
       y se construye a partir de los datos.

3. **Máquinas de vectores soporte con kernels**:
   - Aunque las máquinas de vectores de soporte (SVM) pueden considerarse paramétricas si usan un kernel
    lineal, cuando usan un kernel no lineal, se convierten en un método no paramétrico porque el kernel 
    permite que el modelo sea tan flexible como lo requieran los datos.

4. **Redes neuronales (en algunos casos)**:
   - Aunque las redes neuronales tradicionales tienen un número fijo de parámetros, las redes neuronales profundas con arquitecturas más complejas pueden comportarse como modelos no paramétricos porque aumentan su capacidad a medida que se agrega más información (capas, nodos, datos).

### Comparación con algoritmos paramétricos:
- **Algoritmos paramétricos** (como la regresión lineal, Naive Bayes, o SVM con kernel lineal) asumen una 
estructura funcional fija. Por ejemplo, en una regresión lineal, asumes que la relación entre las variables es lineal, y el modelo solo ajusta los coeficientes de las variables.
- En contraste, los **algoritmos no paramétricos** no asumen esta forma predeterminada, y su estructura 
se define principalmente a partir de los datos.

### Ventajas de los algoritmos no paramétricos:
- **Flexibilidad**: Pueden adaptarse a cualquier estructura de datos sin suposiciones rígidas.
- **Capacidad de capturar relaciones complejas**: Son útiles cuando los datos tienen relaciones no 
lineales o complejas que los modelos paramétricos no pueden capturar.

### Desventajas:
- **Requieren más datos**: Para que funcionen bien y no sobreajusten, suelen necesitar muchos más datos que
los modelos paramétricos.
- **Menor interpretabilidad**: Pueden ser más difíciles de interpretar debido a su complejidad y 
flexibilidad.
- **Mayor costo computacional**: Al no tener una forma funcional definida, pueden requerir mayor tiempo
y memoria, ya que no se simplifican en un conjunto pequeño de parámetros.

En resumen, los algoritmos no paramétricos son ideales para escenarios donde no se conoce de antemano
la estructura de los datos y se desea un modelo flexible que se adapte a los datos observados.
Sin embargo, su flexibilidad viene con el costo de un mayor riesgo de sobreajuste y una necesidad más alta 
de datos.
```

### Breve descripción del Algoritmo KNN

**Definición:** KNN es un algoritmo de clasificación basado en la proximidad a puntos etiquetados. La clasificación de un nuevo punto se determina observando los $K$ puntos más cercanos del conjunto de entrenamiento.

**Concepto básico:** No hay una "fase de entrenamiento" como tal. El algoritmo almacena los datos y, para cada nuevo punto, calcula la distancia a todos los puntos de entrenamiento, seleccionando los $K$ más cercanos para determinar la clase mediante la mayoría de votos

### Breve historia de KNN
El algoritmo k-Nearest Neighbors (k-NN) fue propuesto por Fix y Hodges en 1951 como una técnica no paramétrica para la clasificación, basada en la proximidad entre los puntos. Durante las décadas de 1960 y 1970, el algoritmo comenzó a ganar tracción en áreas como la estadística, la teoría de la información y la inteligencia artificial. En los años 1980 y 1990, con el aumento del poder computacional, k-NN se aplicó ampliamente en reconocimiento de patrones. Aunque algoritmos más avanzados han surgido desde entonces, k-NN sigue siendo relevante por su simplicidad y utilidad, especialmente en problemas con pocos datos.

## FUNDAMENTOS TEORICOS DEL ALGORITMO

Para hacer lo anterior, KNN tiene algunos requisitos:

**Métricas de distancia:**

#### 1. Distancia Euclidiana
$$
d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2} 
$$

**Uso:** 
- Se usa para medir la distancia "real" en espacios euclidianos.
- Ideal para datos numéricos continuos donde se quiere capturar la distancia directa entre puntos.
- Ejemplos de aplicación: clasificación de imágenes, análisis de patrones.

#### 2. Distancia de Manhattan
$$
 d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} |x_i - y_i| 
$$

**Uso:** 
- Utilizada cuando las características están en diferentes dimensiones y la distancia debe ser calculada a lo largo de los ejes.
- Es útil en situaciones donde el movimiento se realiza en una cuadrícula o rejilla.
- Ejemplos de aplicación: problemas en entornos de red, ciudades en un mapa con calles en cuadrícula.

#### 3. Distancia de Minkowski

$$
 d(\mathbf{x}, \mathbf{y}) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{\frac{1}{p}} 
$$

**Uso:** 
- Generaliza las distancias euclidiana y de Manhattan.
- Permite ajustar el parámetro \( p \) para obtener diferentes formas de medir la distancia:
  - \( p = 1 \): distancia de Manhattan.
  - \( p = 2 \): distancia euclidiana.
- Ejemplos de aplicación: problemas que requieren flexibilidad en la medida de distancia.

#### 4. Distancia de Hamming

**Código en Markdown:**
$$
d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} \mathbf{1}_{\{x_i \ne y_i\}} 
$$

**Uso:** 
- Se usa para datos categóricos o binarios, como secuencias o cadenas de caracteres.
- Mide el número de posiciones en las que los elementos correspondientes son diferentes.
- Ejemplos de aplicación: análisis de secuencias de ADN, comparación de cadenas de texto.

### K como Parámetro Crítico

En el algoritmo K-Nearest Neighbors (KNN), el parámetro $K$ representa el **número de vecinos más cercanos que se considerarán para realizar la clasificación o regresión** de un nuevo punto. La elección de $K$ es crucial ya que influye significativamente en el **rendimiento del modelo**. Aquí se detalla su importancia:

 **Efecto del Valor Bajo de K**:
 
Ventajas: Puede capturar patrones locales en los datos, lo cual puede ser útil si el dataset tiene una estructura intrincada.
Desventajas: Un valor muy bajo (por ejemplo, $k=1$) puede llevar a un modelo que se ajuste demasiado a los datos de entrenamiento (overfitting), ya que el modelo puede estar muy influenciado por los puntos de datos cercanos, incluyendo el ruido.

 **Efecto del Valor Alto de K**:
 
Ventajas: Un valor alto de $k$ puede ayudar a suavizar las decisiones y reducir el riesgo de overfitting al considerar un número mayor de vecinos.
Desventajas: Un valor muy alto puede llevar a un modelo que no capture bien las estructuras locales y, en su lugar, promedie demasiado los datos, resultando en subajuste (underfitting).

 **Selección del Valor de K:**
 
Generalmente se utiliza la validación cruzada para seleccionar el mejor valor de $k$. Se prueban diferentes valores y se evalúa el rendimiento del modelo en un conjunto de validación para elegir el que proporcione el mejor equilibrio entre sesgo y varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar el dataset Iris
iris = load_iris()
X = iris.data[:, :2]  # Usar solo las dos primeras características para la visualización
y = iris.target

# Dividir el dataset en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Valores de k a probar
k_values = [1, 3, 5, 8]

# Crear una figura para los gráficos
plt.figure(figsize=(14, 12))

# Entrenar y evaluar el modelo para cada valor de k
for idx, k in enumerate(k_values):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    
    # Crear una malla de puntos para trazar la frontera de decisión
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Graficar la frontera de decisión y los puntos de datos
    plt.subplot(2, 2, idx + 1)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', edgecolor='k', s=50, cmap=plt.cm.RdYlBu)
    plt.title(f'k = {k}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Las gráficas generadas muestran la frontera de decisión del algoritmo KNN para diferentes valores de $k$ (1, 3, 5 y 8) en un conjunto de datos de dos características del dataset Iris. Cada gráfico representa cómo el modelo KNN clasifica los puntos en función de la cercanía a los datos de entrenamiento.

* Puntos de Entrenamiento:
        Los puntos dispersos representan los datos de entrenamiento, con colores que indican a qué clase pertenecen.

* Comparación de k:
        Para $k=1$, la frontera de decisión es muy detallada y se ajusta fuertemente a los datos de entrenamiento, lo que puede indicar sobreajuste.
        Conforme $k$ aumenta (3, 5, 8), la frontera se vuelve más suave, lo que indica que el modelo está generalizando mejor y es menos sensible a ruidos o anomalías.

## **3. Aplicación Matemática y Funcionamiento del Algoritmo**

Vamos a ver cómo funciona KNN paso a paso:

1. **Recoger y almacenar todos los datos de entrenamiento**:  
   Supongamos que tenemos un conjunto de datos con dos clases, A y B, cada una con ciertas características como tamaño y peso.

2. **Calcular la distancia entre el nuevo dato y todos los puntos en el dataset**:  
   Cuando tenemos un nuevo dato, calculamos la distancia entre este punto y todos los demás en el conjunto de datos utilizando la métrica elegida (digamos que usamos la distancia Euclidiana).

3. **Ordenar las distancias**:  
   Luego, ordenamos todos los puntos de entrenamiento en función de qué tan cerca están del nuevo dato.

4. **Seleccionar los "K" vecinos más cercanos**:  
   Elegimos los "K" puntos más cercanos.

5. **Clasificar según la mayoría de votos**:  
   Si estamos haciendo clasificación, vemos cuál es la clase mayoritaria entre los K vecinos. Si la mayoría de los vecinos pertenecen a la clase A, entonces clasificamos el nuevo punto como A.

### K-Nearest Neighbors (KNN) para Regresión

**KNN** también se puede utilizar para **regresión**, donde el objetivo es predecir un valor numérico en lugar de una clase. A continuación, se describe brevemente cómo funciona:

#### Funcionamiento

1. **Seleccionar Vecinos**: Encuentra los $ K $ vecinos más cercanos al punto que se desea predecir.
2. **Calcular la Predicción**: Realiza el promedio de los valores de la variable dependiente (target) de estos \( K \) vecinos.

##### Fórmula para la Predicción

Para un nuevo punto $ x^* $, la predicción $\hat{y}$ es:

$$ \hat{y} = \frac{1}{K} \sum_{i=1}^{K} y_i $$

donde $ y_i $ son los valores de salida de los $K $ vecinos más cercanos.




## Ejemplos Sencillos de Aplicación de KNN

#### 1. Ejemplo Numérico: Clasificación de Puntos en un Plano

**Contexto:**
Supón que tienes tres puntos en un plano 2D, etiquetados como pertenecientes a dos clases: **Clase A** y **Clase B**. Queremos clasificar un nuevo punto usando \( k = 3 \) (consideramos los 3 vecinos más cercanos).

| Punto | Coordenadas (x, y) | Clase  |
|-------|--------------------|--------|
| P1    | (1, 1)             | A      |
| P2    | (2, 2)             | A      |
| P3    | (3, 3)             | B      |
| P4    | (4, 4)             | B      |

El nuevo punto que queremos clasificar es: **(2.5, 2.5)**.

**Paso a paso:**

1. **Calcular distancias (Euclídeas) entre el nuevo punto y los puntos del conjunto de entrenamiento.**
   - Distancia a P1: $ \sqrt{(2.5 - 1)^2 + (2.5 - 1)^2} = \sqrt{2.25 + 2.25} = 2.12 $
   - Distancia a P2: $ \sqrt{(2.5 - 2)^2 + (2.5 - 2)^2} = \sqrt{0.25 + 0.25} = 0.71 $
   - Distancia a P3: $ \sqrt{(2.5 - 3)^2 + (2.5 - 3)^2} = \sqrt{0.25 + 0.25} = 0.71 $
   - Distancia a P4: $ \sqrt{(2.5 - 4)^2 + (2.5 - 4)^2} = \sqrt{2.25 + 2.25} = 2.12 $

2. **Seleccionar los 3 vecinos más cercanos (los de menor distancia):**
   - P2 (distancia = 0.71, Clase A)
   - P3 (distancia = 0.71, Clase B)
   - P1 (distancia = 2.12, Clase A)

3. **Clasificar según la mayoría de clases de los vecinos cercanos:**
   - Hay 2 puntos de Clase A y 1 de Clase B.
   - **Resultado:** El nuevo punto se clasifica como **Clase A**.

#### 2. Ejemplo No Numérico: Clasificación de Frutas según Colores y Forma (K=3)

**Contexto:**
Supón que tienes una lista de frutas etiquetadas según su color y forma. Queremos clasificar una nueva fruta como **Manzana** o **Naranja**.

| Fruta    | Color    | Forma   | Clase     |
|----------|----------|---------|-----------|
| Fruta 1  | Rojo     | Redonda | Manzana   |
| Fruta 2  | Naranja  | Redonda | Naranja   |
| Fruta 3  | Verde    | Redonda | Manzana   |
| Fruta 4  | Naranja  | Ovalada | Naranja   |

La nueva fruta que queremos clasificar tiene las características:
- **Color: Naranja**
- **Forma: Redonda**

**Paso a paso:**

1. **Comparar las características del nuevo dato con las frutas conocidas.**
   - Con Fruta 1 (Rojo, Redonda): Coincide en forma, pero no en color.
   - Con Fruta 2 (Naranja, Redonda): Coincide en color y forma.
   - Con Fruta 3 (Verde, Redonda): Coincide en forma, pero no en color.
   - Con Fruta 4 (Naranja, Ovalada): Coincide en color, pero no en forma.

2. **Seleccionar los 3 vecinos más cercanos (según la coincidencia en color y forma):**
   - Fruta 2: Coincide en ambas características (más cercana).
   - Fruta 1 y Fruta 3: Coinciden solo en la forma (menos cercanas).

3. **Clasificar según la mayoría de clases de los vecinos cercanos:**
   - Fruta 2 es **Naranja**.
   - Fruta 1 es **Manzana**.
   - Fruta 3 es **Manzana**.

4. **Resultado:**
   - Hay 2 Manzanas y 1 Naranja, pero la más cercana es una Naranja, así que dependiendo de la importancia de la coincidencia exacta en color y forma, **puedes clasificar la fruta como Naranja**.

**Comentario:**

Aunque hay 2 frutas clasificadas como **Manzana** que son cercanas en términos de forma (redonda), la fruta más cercana en términos de coincidencia exacta de características es la **Fruta 2**, que es una **Naranja**. En KNN, la coincidencia exacta en ambas características (color y forma) puede tener más peso que la coincidencia en una sola característica, especialmente cuando se trata de una coincidencia perfecta. Por lo tanto, en este caso, se decide clasificar la nueva fruta como **Naranja** debido a la mayor proximidad en características a la **Fruta 2**.

## **4. Mejoras y Optimizaciones de KNN**

### Excepciones en la Clasificación con KNN y el Impacto de Elegir \( k \) Impar

#### Excepciones en la Clasificación

1. **Datos Solapados:**
   - **Descripción:** Cuando las clases se solapan, es decir, cuando las características de diferentes clases se mezclan, puede ser difícil clasificar el nuevo elemento de manera precisa. Por ejemplo, si dos clases tienen áreas de características que se superponen significativamente, los puntos vecinos pueden ser de ambas clases.
   - **Impacto:** En estos casos, el modelo KNN puede tener dificultades para asignar una clase correcta debido a la ambigüedad en la frontera de decisión.

2. **Ruidos y Outliers:**
   - **Descripción:** Si el conjunto de datos contiene ruido (datos incorrectos o irrelevantes) o puntos atípicos (outliers) que no representan bien las clases, estos pueden afectar la clasificación.
   - **Impacto:** Los puntos atípicos pueden influir en la decisión final si están cerca del nuevo punto, distorsionando la clasificación.

3. **Desbalance de Clases:**
   - **Descripción:** Cuando algunas clases están sobrerrepresentadas en el conjunto de datos (es decir, algunas clases tienen muchos más ejemplos que otras), los vecinos cercanos pueden inclinarse hacia la clase mayoritaria.
   - **Impacto:** Esto puede llevar a una clasificación sesgada hacia la clase más frecuente, incluso si el nuevo punto podría pertenecer a una clase minoritaria.

4. **Características Irrelevantes:**
   - **Descripción:** Si las características utilizadas para la clasificación no son relevantes para la separación entre clases, pueden hacer que la clasificación sea inexacta.
   - **Impacto:** Las características irrelevantes pueden hacer que los cálculos de distancia no reflejen adecuadamente la similitud entre puntos.

5. **Escalado de Características:**
   - **Descripción:** Las características con escalas muy diferentes pueden dominar el cálculo de distancias.
   - **Impacto:** Es importante normalizar o estandarizar las características para que todas tengan igual importancia en la distancia calculada.

## Impacto de Elegir $k$ Impar

- **Evitar Empates:**
  - **Descripción:** Elegir un valor impar para \( k \) ayuda a evitar empates en la clasificación. Si \( k \) es impar, siempre habrá una mayoría clara entre los vecinos más cercanos.
  - **Impacto:** Esto es especialmente útil en problemas de clasificación donde se desea una decisión clara y definitiva. Con \( k \) impar, no habrá un empate exacto entre clases, lo que facilita la toma de decisiones.

- **Controlar el Sobreajuste:**
  - **Descripción:** Valores bajos de $k$ (como $k = 1 )$ pueden llevar al sobreajuste, donde el modelo se ajusta demasiado a los datos de entrenamiento y no generaliza bien.
  - **Impacto:** Elegir un $k$ impar y relativamente pequeño puede mejorar la precisión si hay un buen equilibrio entre ajuste y generalización. Sin embargo, $k$ demasiado pequeño puede hacer el modelo sensible a los datos ruidosos. Valores mayores de \( k \) tienden a suavizar la frontera de decisión y reducir la sensibilidad al ruido, pero también pueden perder detalles importantes si \( k \) es demasiado grande.


#### Escalar Datos para Mejorar el Algoritmo KNN

*¿Por qué Escalar los Datos?*

1. **Distancia No Equitativa:**
   - KNN utiliza distancias para determinar la similitud entre puntos. Si las características tienen escalas diferentes (por ejemplo, una característica en metros y otra en centímetros), las características con escalas mayores dominarán la medida de distancia, afectando la clasificación.

2. **Mejora en la Precisión:**
   - Al escalar los datos, todas las características tienen una influencia equitativa en el cálculo de distancias, lo que puede mejorar la precisión del modelo.

Métodos Comunes de Escalado*

1. **Normalización (Min-Max Scaling):**
   - **Descripción:** Escala las características para que estén en un rango específico, normalmente entre 0 y 1.
   - **Fórmula:** $$ X' = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}} $$
   - **Uso:** Útil cuando se necesita que las características estén en un rango específico.

2. **Estandarización (Z-score Normalization):**
   - **Descripción:** Escala las características para que tengan una media de 0 y una desviación estándar de 1.
   - **Fórmula:** $$ X' = \frac{X - \mu}{\sigma}, $$ donde $\mu$ es la media y $\sigma$ es la desviación estándar.
   - **Uso:** Útil cuando los datos tienen una distribución normal o cuando se desea que todas las características contribuyan igualmente.


**Optimización de Cálculo de Distancia**  
Para grandes conjuntos de datos, calcular todas las distancias puede ser costoso. Podemos usar estructuras como los **KD-Trees** o **Ball Trees** para acelerar el proceso de búsqueda de vecinos cercanos.



#### Construcción del KD-Tree

### Pasos del Algoritmo KD-Trees

1. **Inicialización:**
   - Comienza con un conjunto de puntos en un espacio k-dimensional.

2. **Elegir el Eje de División:**
   - Alterna entre los ejes en cada nivel del árbol (por ejemplo, eje x en el nivel 0, eje y en el nivel 1).

3. **Dividir los Datos:**
   - Ordena los puntos según el eje de división actual.
   - Elige el punto en el medio de la lista ordenada como nodo del árbol.

4. **Construir Subárboles:**
   - Divide los puntos en dos grupos basados en el punto de división:
     - **Grupo Izquierdo:** Puntos con valores menores o iguales.
     - **Grupo Derecho:** Puntos con valores mayores.
   - Recurre a los grupos izquierdo y derecho para construir los subárboles.

5. **Repetir el Proceso:**
   - Repite los pasos 2-4 para cada subárbol hasta que todos los puntos se asignen a un nodo del árbol.

6. **Finalizar:**
   - El árbol está completo cuando todos los puntos están asignados y no se pueden dividir más.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

class KDTreeNode:
    def __init__(self, point, left=None, right=None):
        self.point = point
        self.left = left
        self.right = right

def build_kd_tree(points, depth=0):
    if not points:
        return None
    
    k = 2
    axis = depth % k
    points = sorted(points, key=lambda x: x[axis])
    median = len(points) // 2
    
    return KDTreeNode(
        point=points[median],
        left=build_kd_tree(points[:median], depth + 1),
        right=build_kd_tree(points[median + 1:], depth + 1)
    )

def plot_kd_tree(node, depth=0, x_min=0, x_max=10, y_min=0, y_max=10, ax=None):
    if node is None:
        return
    
    k = 2
    axis = depth % k
    if axis == 0:  # Vertical line
        line_x = [node.point[0], node.point[0]]
        line_y = [y_min, y_max]
        ax.plot(line_x, line_y, 'r--', lw=2)
        plot_kd_tree(node.left, depth + 1, x_min, node.point[0], y_min, y_max, ax)
        plot_kd_tree(node.right, depth + 1, node.point[0], x_max, y_min, y_max, ax)
    else:  # Horizontal line
        line_x = [x_min, x_max]
        line_y = [node.point[1], node.point[1]]
        ax.plot(line_x, line_y, 'b--', lw=2)
        plot_kd_tree(node.left, depth + 1, x_min, x_max, y_min, node.point[1], ax)
        plot_kd_tree(node.right, depth + 1, x_min, x_max, node.point[1], y_max, ax)
    
    ax.scatter(*node.point, color='black')
    
def visualize_kd_tree(points):
    root = build_kd_tree(points)
    fig, ax = plt.subplots(figsize=(8, 8))
    plot_kd_tree(root, ax=ax)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('KD-Tree Visualization')
    plt.show()

# Define points
points = [(4, 7), (2, 6), (9, 7), (6, 4), (7, 1), (3, 8), (8, 5), (5, 9), (1, 3)]

# Visualize KD-Tree
visualize_kd_tree(points)
```

Eficiencia del KD-Tree

    Reducción de la Complejidad Computacional:
        Cálculo Exhaustivo: En un espacio k-dimensional, calcular la distancia entre un punto de consulta y todos los puntos en el conjunto de datos tiene una complejidad de $O(n)$, donde nn es el número de puntos en el conjunto de datos.
        KD-Tree: Utilizando un KD-Tree, la complejidad de la búsqueda de vecinos cercanos puede ser reducida a O(log⁡n)O(logn) en el mejor de los casos, donde la estructura del árbol permite descartar grandes regiones del espacio de búsqueda que no contienen puntos cercanos.

    División del Espacio de Búsqueda:
        Un KD-Tree divide el espacio de búsqueda en regiones más pequeñas y maneja la búsqueda de manera eficiente. Cada nodo del árbol divide el espacio en dos mitades a lo largo de un eje, lo que reduce significativamente la cantidad de puntos que deben ser examinados para encontrar los vecinos cercanos.

### Ball Trees

Un **Ball Tree** es una estructura de datos utilizada para encontrar puntos cercanos en un conjunto de datos de manera eficiente. En lugar de dividir el espacio en regiones rectangulares, como lo hace un KD-Tree, un Ball Tree usa esferas o "bolas" para agrupar puntos. Esta técnica es útil especialmente cuando los datos están en un espacio con muchas dimensiones.

####  Cómo Funciona un Ball Tree

1. **División en Bolas:**
   - Agrupamos puntos en esferas que los cubren.
   - Cada esfera tiene un centro (un punto promedio) y un radio (la distancia desde el centro hasta el punto más lejano dentro de la esfera).

2. **Construcción del Árbol:**
   - **1 :** Calculamos el centro y el radio de la esfera que cubre todos los puntos en el nodo actual.
   - **2 :** Dividimos esta esfera en dos grupos, creando esferas más pequeñas para cada grupo.
   - **3 :** Asignamos estas esferas como nodos hijos y repetimos el proceso para cada nodo hijo.
   - **4 :** Continuamos dividiendo hasta que cada nodo tenga un número pequeño de puntos o se alcance una profundidad deseada.

3. **Búsqueda de Vecinos Cercanos:**
   - **1 :** Comenzamos desde la esfera raíz.
   - **2 :** Buscamos en la esfera más cercana a la consulta.
   - **3 :** Si una esfera es demasiado lejana para contener puntos cercanos (basado en la distancia al vecino más cercano encontrado hasta ahora), la ignoramos.
   - **4 :** Repetimos el proceso para los nodos hijos de las esferas que no fueron ignoradas.

4. **Finalización:**
   - Cuando todos los nodos relevantes han sido explorados, hemos encontrado los vecinos más cercano.

La validación cruzada (cross-validation) es una técnica utilizada en el aprendizaje automático para evaluar el rendimiento de un modelo y evitar problemas como el sobreajuste (overfitting) o el subajuste (underfitting). El objetivo es obtener una estimación más robusta de la capacidad del modelo para generalizar a datos nuevos.
¿Qué es la Validación Cruzada?

La validación cruzada es un método que divide el conjunto de datos en varias partes (o "folds") para evaluar el modelo de manera más fiable. En lugar de usar solo una división del conjunto de datos en entrenamiento y prueba, la validación cruzada utiliza múltiples particiones para probar el modelo en diferentes subconjuntos del conjunto de datos.
Proceso de Validación Cruzada

    División del Conjunto de Datos:
        El conjunto de datos se divide en kk partes (o "folds"). Cada parte es utilizada tanto para entrenamiento como para prueba en diferentes iteraciones.

    Entrenamiento y Evaluación:
        En cada iteración, se selecciona uno de los kk folds como conjunto de prueba y el resto como conjunto de entrenamiento.
        El modelo se entrena en el conjunto de entrenamiento y se evalúa en el conjunto de prueba.
        Este proceso se repite kk veces, cada vez con un fold diferente como conjunto de prueba.

    Promedio de Resultados:
        Se calculan las métricas de rendimiento (como precisión, exactitud, F1-score) para cada una de las kk iteraciones.
        El resultado final es el promedio de estas métricas, proporcionando una estimación más robusta del rendimiento del modelo.
